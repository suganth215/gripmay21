```python
df.loc[df['winner'] == 'Rising Pune Supergiants', 'winner'] = 'Rising Pune Supergiant'
```

```python
wins_dict=df["winner"].value_counts()
wins_dict.to_dict()
teams_list=list(wins_dict.keys())
teams_list
no_of_wins=list(wins_dict.values)
no_of_wins
```

```python
plt.figure(figsize=(10,7))
plt.barh(teams_list,no_of_wins,color=['#3769a4','#eece00','#412a78','#f4141c','#ee3642','#2c5bb1','#304680','#ef712f','#294464','#bf427e','#fb4825','black','#407fd3','#f97443'])
plt.title('Number of wins of each team')
plt.ylabel('Number of wins')
plt.xlabel('Teams')
for index, value in enumerate(no_of_wins):
    plt.text(value, index, str(value))
plt.show()
```

>Plotting the Horizontal bar graph of the number of wins of each team
